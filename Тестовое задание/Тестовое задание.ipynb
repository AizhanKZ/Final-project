```python
#8
import pandas as pd
from scipy.stats import ttest_ind

df = pd.read_excel('Данные для тестового задания.xlsx', sheet_name=1)
results = []

for exp in df['experiment_num'].unique():
    sub = df[df['experiment_num'] == exp]
    
    control = sub[sub['experiment_group'] == 'control']['revenue']
    test = sub[sub['experiment_group'] == 'test']['revenue']
    
    control_arpu = control.mean()
    test_arpu = test.mean()
    
    t_stat, p_value = ttest_ind(test, control, equal_var=False)
    
    uplift = (test_arpu - control_arpu) / control_arpu * 100
    
    results.append([
        exp,
        control_arpu,
        test_arpu,
        uplift,
        p_value
    ])

pd.DataFrame(results, columns=[
    "experiment_num",
    "control_arpu",
    "test_arpu",
    "uplift_%",
    "p_value"
])
```

```python
#18
from statsmodels.stats.proportion import proportions_ztest

successes = [1003, 1099]  
samples = [100047501, 100001055] 

stat, p_value = proportions_ztest(successes, samples)

cr_a = successes[0] / samples[0]
cr_b = successes[1] / samples[1]

uplift = (cr_b - cr_a) / cr_a * 100

print("CR A:", cr_a)
print("CR B:", cr_b)
print("Uplift %:", uplift)
print("p-value:", p_value)
```

```python
if p_value < 0.05:
    if cr_b > cr_a:
        print("Вывод: B статистически значимо лучше → внедряем")
    else:
        print("Вывод: B статистически значимо хуже → не внедряем")
else:
    print("Вывод: нет статистически значимого различия")
```
